Function named countTrips. It will take a pandas bike trips history dataframe as
its input. The function should return an integer indicating the number of trips that are
contained in the dataframe.

Function named avgTripDuration that will return the average trip duration (in
seconds) for a given pandas bike trips history dataframe.

```python
def avgTripDuration(df):
    #errors='coerce' to handle problematic date formats by converting them to NaT
    df['started_at'] = pd.to_datetime(df['started_at'], errors='coerce')
    df['ended_at'] = pd.to_datetime(df['ended_at'], errors='coerce')
    #calculate trip duration only for rows where both start and end times are valid
    df['tripduration'] = (df['ended_at'] - df['started_at']).dt.total_seconds()
    #drop NaN values before calculating mean
    return df['tripduration'].dropna().mean()

avgTripDuration(df)
```

Function named countAllStations that will return the number of unique stations
(across both start and end stations) in a given bike trips history dataframe.

```python
def countAllStations(df):
    #return the int number of unique stations (across both start and end stations) in bike trips history dataframe
    unique_start = df['started_at'].unique()
    unique_end = df['ended_at'].unique()
    all_unique = set(unique_start).union(set(unique_end))
    return len(all_unique)

print(countAllStations(df))
```

Function named avgTripsByDayOfWeek that will return a 7
element list containing the average number of trips taken on each day of the week. 

```python
def avgTripsByDayOfWeek(df):
    
    #returns a 7-element list containing the average number of trips taken on each day of the week.
    #The first entry is for Sunday, the last for Saturday, using 'started_at' as trip start time. 
    
    #ensure 'started_at' is datetime
    df['started_at'] = pd.to_datetime(df['started_at'])
    #add day of week (where Monday=0, Sunday=6)
    df['day_of_week'] = df['started_at'].dt.dayofweek
    #add 'trip_date' column for calendar day grouping
    df['trip_date'] = df['started_at'].dt.date
    #count trips per day per weekday
    trips_per_day = df.groupby(['trip_date', 'day_of_week']).size().reset_index(name='trips_count')
    #average trips by each weekday
    avg_by_day = trips_per_day.groupby('day_of_week')['trips_count'].mean()
    #arrange so Sunday is first: [Sunday, Monday, ..., Saturday]
    avg_list = [avg_by_day.get(6, 0)] + [avg_by_day.get(i, 0) for i in range(0, 6)]
    return avg_list
    
print("Average trips by day of week [from Sunday to Saturday respectively]:",avgTripsByDayOfWeek(df))
```

Function named topXdepartures that will find the x stations that have the most 
departures, sorted in descending order. If there are ties, include those as well. 

```python
import pandas as pd

def get_column(df, possible_names):
    """
    Returns the first column name from 'possible_names' that exists in the dataframe.
    Raises an error if none are found.
    """
    for name in possible_names:
        if name in df.columns:
            return name
    raise KeyError(f"None of the expected columns found: {possible_names}")


def topXdepartures(x, df):
    """
    Finds the top X stations with the most departures.
    Includes tied stations 

    Returns a dataframe with:
    -rank (1 = highest departures, ties share rank)
    -start station id
    -number of departures

    Inputs:
      -x: number of top ranks to return (ties included)
      -df: CitiBike trip history dataframe
    """

    #identify the correct start station id column
    #tries multiple possible versions that might appear in CitiBike datasets
    start_col = get_column(df, [
        "start station id",
        "Start Station ID",
        "start_station_id",
        "start id",
        "start_id"
    ])

    #count the number of departures per station
    #groupby → groups by station ID
    #size()  → counts how many rows per station
    counts = (
        df.groupby(start_col)
          .size()
          .reset_index(name="number of departures")
    )

    #sort stations by number of departures in descending order
    counts = counts.sort_values("number of departures", ascending=False)

    #assign dense ranks:
    #- highest departure count = rank 1
    #- ties get the same rank
    #- next unique value gets rank +1 (no gaps)
    counts["rank"] = counts["number of departures"].rank(
        method="dense",   # ensures no skipped ranks
        ascending=False
    ).astype(int)

    #extract all unique ranks, sorted (1, 2, 3, ...)
    unique_ranks_sorted = sorted(counts["rank"].unique())

    #determine the cutoff rank:
    #if x = 10, we take the 10th rank.
    #but if x > total ranks, take the last available rank.
    cutoff_rank = unique_ranks_sorted[min(x, len(unique_ranks_sorted)) - 1]

    #filter stations whose rank is within the cutoff (ties included)
    result = counts[counts["rank"] <= cutoff_rank]

    #return the dataframe with correct column order
    return result[["rank", start_col, "number of departures"]]


print(topXdepartures(3, df))
```
